# Web scraping coronavirus data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np


url="https://www.worldometers.info/coronavirus/#countries"

r=requests.get(url)
soup=BeautifulSoup(r.content,'html.parser')

result=soup.find('table',id="main_table_countries_today")
en=result.find('tbody')
cen=en.find_all('td')

for entries in cen:
    u=entries.text.strip()
    print(u,end=" ")
```

#  Scraped Data in Excel CSV Format 

Before inserting data in Excel CSV, The data is redesign

```python
li=[]
listy = [[] for i in range(225)]
cnt=0;k=0

for entries in cen:
    u=entries.text.strip()
    if (u=="World"):
        k=1
    if (k==1):
        li.append(u)

        
cnt=0;ind=1;flg=0

for i in li:
    cnt=cnt+1
    if (flg==0):
        listy[0].append(i)
        if (cnt==14):
            flg=1;cnt=0
    
    if (flg==1):
        if cnt <=8:
            continue
        listy[ind].append(i)
        if cnt==22:
            cnt=0
            ind=ind+1
        

    
df = pd.DataFrame(listy, columns =['Country','Total Cases','New Cases','Total Death','New Deaths','Total Recovered','other','Active Cases','Serious Critical','Tot Cases/1M pop','Deaths/1M pop','Total Tests','Tests/1M pop','Population'])
print(df)
df.to_csv('Covid.csv')
```
